```python
import nltk
import re
import string
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('punct')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
corpus = """
In 2024, Microsoft launched several AI-powered tools to help students learn faster and smarter. 
Many universities in India and the United States started using these platforms to improve online education. 
Students like Rahul Sharma from Delhi said that artificial intelligence helped him understand complex topics more easily. 
However, some experts at Stanford University warned that overdependence on technology could reduce critical thinking skills.
"""
```

```python
sentence = sent_tokenize(corpus)
```

```python
print(sentence)
```

```python
sentence = corpus.lower()
```

```python
sentence = re.sub(r'\d+', '', corpus)
```

```python
sentence = corpus.translate(str.maketrans('', '', string.punctuation))
```

```python
words = word_tokenize(sentence)
print(words)
```
