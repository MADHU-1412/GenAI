#### Sentence Transformers on Hugging face

pything framework for state-of-art sentence, text and image embeddings one of the model used in huggingface embeddings class. we have also added an alias SentenceTransformers Embeddings for users who are more familiar with directly using that package

```python
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
embeddings = HuggingFaceEmbeddings(model_name= "all-MiniLM-L6-v2")
```

```python
text = "this is the text document"
query_res = embeddings.embed_query(text)
```
