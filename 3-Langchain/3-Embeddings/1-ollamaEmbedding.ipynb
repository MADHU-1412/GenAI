
## ollama Embedding

ollama supports embedding, making possible to build RAG application that combinetext prompt with existing document and other data 

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings= (OllamaEmbeddings(model="gemma:2b"))
```

```python
embeddings
```

```python
r1 = embeddings.embed_documents(
    [
        "Alpha is First letter of Greek Alphabet",
        "Beta is Second letter of Greek Aphabet",
    ]
)
```

```python
r1
```

```python
embeddings.embed_query("What is the second Alphabet in Greek")
```
