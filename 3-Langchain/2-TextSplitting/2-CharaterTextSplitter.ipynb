# Character Text splitter

Simpliest method. split based on chrcter sequence, which default to '\n\n'. chunk length is measure by number of characters

    1. how test is split : by single character splitters
    2. chunk size is measured: by the number of characters

```python
from langchain_community.document_loaders import TextLoader
loader = TextLoader('speech.txt')
doc = loader.load()
doc
```

```python
#text splitter
from langchain_text_splitters import CharacterTextSplitter
text_splitter = CharacterTextSplitter(separator='\n\n', chunk_size= 100, chunk_overlap = 20 )
text_splitter.split_documents(doc)
```

```python
speech = ""
with open('speech.txt') as f:
    speech = f.read()

text_splitter = CharacterTextSplitter(chunk_size = 100, chunk_overlap = 20)
text = text_splitter.create_documents([speech])
```
