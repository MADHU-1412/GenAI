## ChromaDB

#### AI-Native open_source vector database focused on developer productivity and happiness. chroma db is licenced under apachee

```python
#Building ssample vectorDB

#importing lib
from langchain_chroma import Chroma
from langchain_community.document_loaders import TextLoader
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
#text loading
loader = TextLoader("speech.txt")
data = loader.load()
```

```python
data
```

```python
#splliter
text_splitter = RecursiveCharacterTextSplitter(chunk_size = 100, chunk_overlap = 20)
split = text_splitter.split_documents(data)
```

```python
#embeddings

embeddings = OllamaEmbeddings()
vectordb = Chroma.from_documents(documents=split, embedding= embeddings)
vectordb
```

```python
# query
query = "what is Generative AI?"
docs = vectordb.similarity_search(query)
docs[0].page_content
```
